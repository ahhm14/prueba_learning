En IPython, podemos crear gráficos sencillos y rápidos para obtener insigths importantes

Si queremos saber cuantas personas sobrevivieron y cuantas no, podemos plotear un gráfico de barras utilizando la librería Pandas o Seaborn

```python
pd.value_counts(train_data['Survived']).plot.bar()
;
```

```python
sb.countplot('Survived',data=train_data)
;
```

Del gráfico anterior queda claro que no muchas personas sobrevivieron.

De las 891 personas en el primer conjunto de datos, solo 350, el 38.4% del conjunto de datos de entrenamiento sobrevivió

Podemos clasificar y agrupar nuestros datos en función a diferentes variables

```python
train_data.groupby(['Sex', 'Survived'])['Survived'].count()
```

A continuación, graficaremos los datos en relación al sexo y el índice de supervivencia 

```python
train_data[['Sex','Survived']].groupby(['Sex']).mean().plot.bar()
sb.countplot('Sex',hue='Survived',data=train_data,)
plt.show()
```

Ahora graficaremos el índice de supervivencia en relación a la clase de boleto

```python
sb.countplot('Pclass', hue='Survived', data=train_data)
plt.title('Pclass: Sruvived vs Dead')
plt.show()
```

Ahora relacionaremos ambas variables en una matriz

```python
pd.crosstab([train_data.Sex,train_data.Survived],train_data.Pclass,margins=True)
```

Podemos graficar la matriz con el método de Pandas ' factorplot ' 

```python
sb.factorplot('Pclass', 'Survived', hue='Sex', data=train_data)
plt.show()
```

Si queremos conocer algunos datos estadísticos en relación a las edades:

```python
print('La persona más vieja en sobrevivir tenía:',train_data['Age'].max(),'años')
print('La persona más joven en sobrevivir tenía:',train_data['Age'].min(),'años')
print('El promedio de la edad de los sobrevivientes fue de:',train_data['Age'].mean(),'años')
```

```python
train_data['Survived'].mean()
```

mean: función promedio

Un método muy importante de Python es el comando  ' groupby ' ya que agrupa variables con valores iguales

```python
train_data.groupby(['Sex']).mean()
```

```python
train_data.groupby(['Sex', 'Pclass']).mean()
```

También, podemos utilizar condicionales para desplegar segmentos de datos más específicos

```python
train_data[train_data['Age']>18].groupby(['Sex','Pclass']).mean()
```
